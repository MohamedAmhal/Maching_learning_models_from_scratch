# Support Vector Machine (SVM)

The support vectors machines are the best classifiers in machine learning , the basics of this model is to maximaze the distance between classes (margin)
it exist a lot formula to seperate the data (linear , not linear, kernel...)
the SVM has the possibility to separate the linear data and also the no linear data .

**Advantages of SVMs**:

- Highly effective in high-dimensional spaces.
- Efficient when the dimensionality of the space is larger than the number of training samples.
- or decision-making, they do not use all training samples but only a subset (support vectors). As a result, these algorithms require less memory.
    
**Disadvantages**:

- If the number of attributes is much greater than the number of samples, the performance is not as good.
- Since they are classification methods, they do not provide probability estimates.







## **1. The data set** :

The Iris dataset, also known as Fisher's Iris or Anderson's Iris, was introduced in 1936 by Ronald Fisher in his paper "The Use of Multiple Measurements in Taxonomic Problems." It serves as an example of the application of linear discriminant analysis. This dataset consists of 50 samples for each of the three iris species (Iris setosa, Iris virginica, and Iris versicolor). Four characteristics were measured for each sample: the length and width of the sepals and petals, expressed in centimeters. Based on the combination of these four variables, Fisher developed a linear discriminant analysis model to distinguish between the species.

In Scikit-learn, SVMs are implemented in the sklearn.svm module. In this case, we will focus on the linear version (Scikit-learn uses the libLinear and libSVM libraries). We will use the Iris dataset, but to simplify the visualization, we will only use the first two features (sepal length and sepal width).



## **2. Importing the libiraries** :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
#ignore the warning related to convergence :
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)
```

```python
# Chargement des données
iris = datasets.load_iris()
```

```python
X, y = iris.data[:, :2], iris.target
# On conserve 50% du jeu de données pour l'évaluation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50)
```

```python
# test 
print('X_train shape :', X_train.shape)
print('X_test shape :', X_test.shape)
print('y_train shape :', y_train.shape)
print('y_test shape :', y_test.shape)
```

## **3. The model SVM**:

```python
#training the model : 
#regularization 
C = 1.0
classifier = svm.LinearSVC(C=C)
classifier.fit(X_train, y_train)
```

```python
score = classifier.score(X_test, y_test)
print("The score of the model is :", score)
```

## **4. The visualisation**: 

```python
# Create the decison surface :
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
# for displaying the surface of the decison boundry :
h = max((x_max - x_min) / 100, (y_max - y_min) / 100)
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max,h))
# the decision space :
Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
# desplaying the points of the model SVM 
plt.scatter(X_train[:, 0], X_train[:, 1], label="train", edgecolors='k',
c=y_train, cmap=plt.cm.coolwarm)
plt.scatter(X_test[:, 0], X_test[:, 1], label="test", marker='*', c=y_test,
cmap=plt.cm.coolwarm)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title("LinearSVC")
plt.show()
```

The classifier produces linear decision boundaries. This is sufficient to separate one of the three classes from the other two. However, when only considering the first two attributes (sepal length and sepal width), the other two classes do not appear to be linearly separable.

To improve the classification, one could either:

- Use a non-linear model that can better capture the complex boundaries between the classes.
- Add more attributes, hoping that the additional information will help to linearly separate the remaining two classes.


These approaches are common strategies when dealing with datasets where linear separation is not achievable with the existing features.

```python
lin_svc = svm.LinearSVC(C=C).fit(X_train, y_train)
svc = svm.SVC(kernel='linear', C=C).fit(X_train, y_train)
titles = ['SVC with linear kernel', 'LinearSVC (linear kernel)']
fig = plt.figure(figsize=(12, 4.5))
for i, clf in enumerate((svc, lin_svc)):
    plt.subplot(1, 2, i + 1)
    plt.subplots_adjust(wspace=0.4, hspace=0.4)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X_train[:, 0], X_train[:, 1], label="train", edgecolors='k', c=y_train, cmap=plt.cm.coolwarm)
    plt.scatter(X_test[:, 0], X_test[:, 1], label="test", marker='*', c=y_test, cmap=plt.cm.coolwarm)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.title(titles[i])
plt.show()
```

## **5. Second data set : (digits)**

this data contient the images related to digits (0 --> 9) , the data exists in the libiraries sklesrn


```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
X , y =  digits.data, digits.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

c_value = [0.001, 0.01, 0.1, 1, 10, 100]

best_score = 0
best_c = 0

for c in c_value:
    classifier = svm.LinearSVC(C= c)
    classifier.fit(X_train, y_train)
    score = classifier.score(X_test, y_test)
    print('c = ', c, ' score = ', score)
    if score > best_score:
        best_score = score 
        best_c = c


print('the best parameter c is equal to :', best_c, '  and the final score for this parameter is equal to : ', best_score)
```

#### using the methid GridSearch in sklearn :

```python
from sklearn.model_selection import GridSearchCV
```

```python
input = {'C' : [0.001, 0.01, 0.1, 1, 10, 100]}

lr = svm.LinearSVC()

grid_search = GridSearchCV(lr, input)

grid_search.fit(X_train, y_train)


best_params = grid_search.best_params_
score = grid_search.score(X_test,  y_test)


print(f'Best parameters: {best_params}')
print(f'Score for best parameters: {score}')
```
